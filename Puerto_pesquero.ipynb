```python
# Importar librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo para gráficos
sns.set(style="whitegrid")
```

# Paso 2: Análisis del Emplazamiento

```python
# Datos ficticios de embarcaciones y profundidades
embarcaciones = pd.DataFrame({
    'Tipo': ['Pequeña', 'Mediana', 'Grande'],
    'Calado_m': [1.2, 1.8, 2.5],
    'Capacidad_Toneladas': [2, 10, 25]
})

print(embarcaciones)

# Gráfico de dispersión para visualizar la relación entre tipo de embarcación y calado
plt.figure(figsize=(10, 6))
sns.scatterplot(data=embarcaciones, x='Tipo', y='Calado_m', size='Capacidad_Toneladas', legend=False, sizes=(100, 500))
plt.title('Relación entre Tipo de Embarcación y Calado')
plt.xlabel('Tipo de Embarcación')
plt.ylabel('Calado (m)')
plt.show()
```

# Paso 3: Indicadores de Sostenibilidad

```python
# Datos ficticios de indicadores de sostenibilidad
indicadores = pd.DataFrame({
    'Indicador': ['Rentabilidad', 'Emisiones de CO2', 'Creación de Empleo'],
    'Económico': [8, 3, 7],
    'Ambiental': [2, 9, 4],
    'Social': [6, 4, 8]
})

print(indicadores)

# Gráfico de radar para visualizar los indicadores de sostenibilidad
labels = indicadores['Indicador']
num_vars = len(labels)

# Crear el ángulo de cada categoría en el gráfico de radar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Repetir el primer valor para cerrar el círculo
values = indicadores[['Económico', 'Ambiental', 'Social']].values.tolist()
values = [v + [v[0]] for v in values]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values[0], color='r', alpha=0.25, label='Económico')
ax.fill(angles, values[1], color='g', alpha=0.25, label='Ambiental')
ax.fill(angles, values[2], color='b', alpha=0.25, label='Social')

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.legend(loc='upper right')
plt.title('Indicadores de Sostenibilidad')
plt.show()
```

# Paso 4: Análisis de Datos y Ciencia de Datos

```python
# Datos ficticios de tráfico portuario
tráfico = pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'],
    'Entradas': [120, 150, 170, 200, 180, 210],
    'Salidas': [110, 140, 160, 190, 175, 205]
})

print(tráfico)

# Gráfico de líneas para visualizar el tráfico portuario
plt.figure(figsize=(10, 6))
sns.lineplot(data=tráfico, x='Mes', y='Entradas', marker='o', label='Entradas')
sns.lineplot(data=tráfico, x='Mes', y='Salidas', marker='o', label='Salidas')
plt.title('Tráfico Portuario Mensual')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Barcos')
plt.legend()
plt.show()
```
